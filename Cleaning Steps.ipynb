```python
# Kita menemukan hal menarik setelah melihat sekilas isi dari 5 baris pertama
# colom "Name", "LongName", dan "PlayerUrl" mengandung informasi mengenai nama pemain dan beberapa karakter non huruf
# Hanya colom "PlayerUrl" yang mengandung informasi mengenai nama lengkap pemain yang di sematkan di Url
```

```python
# Kita kemudian bisa mendapatkan Nama Lengkap pemain dari kolom 'playerUrl' 
# dengan cara menghapus karakter yang tidak diinginkan dari kolom tersebut.
```

```python
# Step 1 
# (Menghilangkan/menghapus 31 karakter awal dari masing masing nilai pada kolom "playerUrl")

fifa21['playerUrl'] = fifa21['playerUrl'].str.slice(31)
fifa21['playerUrl']
```

```python
# Step 2
# (menghilangkan karakter "/" di depat data)

fifa21['playerUrl'] = fifa21['playerUrl'].str.lstrip('/')
fifa21['playerUrl']
```

```python
# Step 3 

fifa21['playerUrl'] =  fifa21['playerUrl'].str.replace('\d+', '', regex=True)
fifa21['playerUrl']
```

```python
# Step 4
# (menghilangkan karakter "/" di belakang setiap data dari kolom "playerUrl")

fifa21['playerUrl'] = fifa21['playerUrl'].str.rstrip('/')
fifa21['playerUrl']
```

```python
# Step 5
# (Mengubah setiap awal kata dari nilai dalam kolom "playerUrl" menjadi huruf kapital di setiap awal kata)

fifa21['playerUrl'] = fifa21['playerUrl'].str.title()
fifa21['playerUrl']
```

```python
# Step 6
# (pisahkan setiap kata yang dibatasi dengan karakter "-")

fifa21['playerUrl'].str.split('-', expand=True)
```

```python
# Step 7
# (Menambahkan kolom kedalam dataframe)

fifa21[['name1', 'name2', 'name3', 'name4', 'name5', 'name6']] = fifa21['playerUrl'].str.split('-', expand=True)
fifa21.head()
```

'Height', 'Weight' and 'Preferred Foot'

```python
#Memeriksa columns 'Height','Weight', 'Preferred Foot'

for column in ['Height','Weight', 'Preferred Foot'] :
    values = fifa21[column].unique()
    print(f'{column}-{values}.\n')
```

```python
# menggunakan fifa_df.unique() untuk memeriksa kolom 'Height','Weight', 'Preferred Foot'.
# Kolom 'Height' memiliki dua nilai entri data yang berbeda yaitu '178cm' dan '5'10" (cm & inci).
# Kolom 'Weight' memiliki dua nilai entri data yang berbeda, yaitu '75kg' dan '110lbs'.
# Kolom 'Preferred Foot' juga memiliki dua nilai entri data yang berbeda yaitu 'Kiri' dan 'Kanan'.
# Setelah mengidentifikasi hal ini, nilai kolom 'Height' dan 'Weight' harus dalam ukuran tertentu untuk dapat melakukan analisis lebih lanjut, 
# sementara nilai kolom 'Preferred Foot' tidak masalah.
```

```python
# Membuat fungsi untuk mengubah nilai ke Kg

def all_to_kg(x):
    if 'kg' in x:
        x = x.strip('kg')
        return x
    elif 'lbs' in x:
        x = x.strip('lbs')
        Weight = round((float(x) * 0.45359237), 2)
        return Weight

fifa21['Weight'] = fifa21['Weight'].apply(all_to_kg).astype('int64')
fifa21['Weight'].unique()
```

```python
# Membuat fungsi untuk mengubah nilai ke cm

def all_to_cm(x):
    if "'" and '"' in x:
        feet, inches = x.split("'")
        total_inches = int(feet) * 12 + int(inches[:-1])
        height_cm = total_inches * 2.54
    elif 'cm' in x:
        height_cm = int(x[:-2])
        total_inches = round(height_cm / 2.54)
    else:
        feet, inches = x.split("'")
        total_inches = int(feet) * 12 + int(inches[:-1])
        height_cm = total_inches * 2.54
    return round(height_cm, 2)

fifa21['Height'] = fifa21['Height'].apply(all_to_cm).astype('int64')
fifa21['Height'].unique()
```

```python
# Ubah nama kolom 'Height' dan 'Weight' dengan menambahkan akhiran 'cm & kg'

fifa21 = fifa21.rename(columns = {'Height' : 'Height (cm)', 'Weight' : 'Weight (kg)'})
fifa21[['Height (cm)', 'Weight (kg)']].head(2)
```

```python
# Memeriksa columns 'Height (cm)','Weight (kg)', dan 'Preferred Foot'

fifa21[['Height (cm)', 'Weight (kg)', 'Preferred Foot']].info()
```

```python
# PROSES
    # Membuat fungsi untuk menghapus 'kg dan lbs' di kolom 'Berat' dan mengonversi semua 'lbs' menjadi 'kg' dengan mengalikannya dengan (0,45359237) per kg dan juga mengubah tipe data menjadi int64.
    # Fungsi lain juga didefinisikan untuk membagi, menghapus dan mengubah 'cm dan kaki & inci' menjadi 'cm' dan mengubah tipe data menjadi int64.
    # Column 'Height' dan 'Weight' diubah namanya menjadi 'Height (cm)' dan 'Weight (kg)'
    # Column 'Kaki yang disukai' sudah cukup baik dan tidak disentuh.
```

'Value, 'Wage', 'Release Clause' dan 'Hits'

```python
# Memeriksa columns 'Value, 'Wage', 'Release Clause' dan 'Hits'

for column in fifa21[['Value', 'Wage', 'Release Clause', 'Hits']] :
    values = fifa21[column].unique()
    print(f'{column}-{values}.\n')
```

```python
# fifa_df.unique() untuk memeriksa kolom 'Value, 'Wage', 'Release Clause' dan 'Hits'
# Kolom-kolom yang disebutkan di atas memiliki entri data yang berbeda dengan awalan dan akhiran seperti '€' untuk mata uang euro, 'K' untuk ribuan, 'M' untuk jutaan.
# Kolom 'Hits' juga memiliki masalah entri data seperti akhiran 'K' untuk ribuan serta baris 'nan' yang kosong.
```

```python
# Step 1

fifa21 = fifa21.fillna(value = {'Hits' : '0'})
fifa21['Hits'].sample(5)
```

```python
# Step 3

for column in fifa21[['Value', 'Wage', 'Release Clause', 'Hits']]:
    fifa21[column] = fifa21[column].replace({'K':'*1e3', 'M':'*1e6'}, regex = True).map(pd.eval).astype('int64')
    
fifa21[['Value', 'Wage', 'Release Clause', 'Hits']].sample(5)
```

```python
# Step 4

fifa21 = fifa21.rename(columns = {'Value' : 'Value (€)', 'Wage' : 'Wage (€)', 'Release Clause' : 'Release Clause (€)'})
fifa21[['Value (€)', 'Wage (€)', 'Release Clause (€)']].head(2)
```
